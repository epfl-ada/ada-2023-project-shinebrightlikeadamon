```python
#Merge together IMDB stuff
#starting merging only with 10000 exampls and maybe figure out cleaning also on subset so doesn't load too long, then apply too whole dataset
#name.basics and principals not included for now
imdb_title_akas = pd.read_csv(IMDB_PATH + "title.akas.tsv", sep='\t', nrows= 10000)
imdb_title_basics = pd.read_csv(IMDB_PATH + "title.basics.tsv", sep='\t', nrows= 10000)
imdb_title_crew = pd.read_csv(IMDB_PATH + "title.crew.tsv", sep='\t', nrows= 10000)
#imdb_title_principlas = pd.read_csv(IMDB_PATH + "title.principals.tsv", sep='\t', nrows= 1000)
#imdb_name_basics = pd.read_csv(IMDB_PATH + "name.basics.tsv", sep='\t', nrows= 1000)
imdb_title_ratings = pd.read_csv(IMDB_PATH + "title.ratings.tsv", sep='\t', nrows= 10000)
```

```python
print(imdb_title_akas.shape)
print(imdb_title_basics.shape)
print(imdb_title_crew.shape)
print(imdb_title_ratings.shape)
```

```python
imdb_title_akas = imdb_title_akas.rename(columns={"titleId": "tconst"})
imdb_merged1 = pd.merge(imdb_title_akas, imdb_title_basics, on="tconst")
imdb_merged1 = imdb_merged1.drop(columns=['title'])
print(imdb_merged1.shape)
imdb_merged2 = pd.merge(imdb_merged1, imdb_title_crew, on="tconst")
print(imdb_merged2.shape)
imdb_merged = pd.merge(imdb_merged2, imdb_title_ratings, on="tconst")
print(imdb_merged.shape)
```
